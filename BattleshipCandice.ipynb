## Présentation du jeu 

La bataille navale, appelée aussi touché-coulé, est un jeu de société pour deux joueurs.
Chaque joueur a deux grilles. Sur une grille tenue secrète ils placent leurs bateaux de tailles différentes, en général, ce sont des bateau de dimension 1xn où n appartient {2,3,4,5} positionnés de manière horizontale ou verticale sur la grille. Le gagnant est celui qui parvient à couler (c’est-à-dire toucher toutes les cases de chaque bateau) tous les bateaux de l'adversaire avant que tous les siens ne le soient.  


## Règles

La bataille navale oppose deux joueurs. Chaque joueur dispose de deux grilles carrées de côté 10, dont les lignes sont numérotées de 1 à 10 et les colonnes de A à J, ainsi que d'une flotte composée de quelques bateaux de deux à cinq cases de long.
L'une des grilles représente la zone contenant sa propre flotte. Au début du jeu, chaque joueur place ses bateaux sur sa grille, en s'assurant que deux bateaux ne se superposent pas. L'autre grille représente la zone adverse, où il cherchera à couler les bateaux de son adversaire.
Chaque joueur, à son tour, annonce une case (par exemple « B6 »), et son adversaire lui répond si le tir tombe à l'eau ou au contraire s'il touche un bateau. Dans ce dernier cas, il annonce « touché » s'il reste des cases intactes au bateau ciblé, et « touché-coulé » sinon.

Les croix sur l'image représentent les cases annoncées par l'adversaire.




## Objectifs

- trouver une stratégie pour gagner rapidement
- savoir où possitionner ses bateaux pour qu'ils soient le moins vite touchés
-programmer la bataille navale pour voir si cette stratégie est efficace

```python
"on va chercher à savoir où est-ce que les bateaux, posées aléatoirement, ont le plus de chance d'être positionnés"
"bien sûr quand on a touché un bateau on va attaquer sur les cases autour de celle touchée"
from random import*
import numpy as np
import matplotlib.pyplot as plt

def test_hori(N,pcase_y,pcase_x,L): #si le bateau est horizontal, on vérifie qu'il n'y a pas de chevauchement
    c=0
    for k in range (pcase_x,pcase_x+L):
        if N[k][pcase_y]==0:      
            c+=1
    return c

def test_vert(N,pcase_y,pcase_x,L): #si le bateau est vertical, on vérifie qu'il n'y a pas de chevauchement
    c=0
    for k in range (pcase_y,pcase_y+L):
        if N[pcase_x][k]==0:
            c+=1
    return c

def grille (n,B):
    N=np.zeros((n,n))
    for i in range (len(B)):
        orientation=randint(0,1) #si orientation=0-> horizontal sinon vertical
        L=B[i]
        validite = False
        while not(validite) :   #on vérifie qu'il n'y a pas de chevauchement avec d'autres bateaux
            pcase_x=randint(0,n-1) #on prend deux nombres entre 1 et n-1
            pcase_y=randint(0,n-1)
            if orientation ==0 and pcase_x<=n-L and test_hori(N,pcase_y,pcase_x,L)==L:
                for k in range (pcase_x,pcase_x+L):
                    N[k][pcase_y]=1
                validite = True
            elif orientation ==1 and pcase_y<=n-L and test_vert(N,pcase_y,pcase_x,L)==L:
                for k in range (pcase_y,pcase_y+L):
                    N[pcase_x][k]=1
                validite = True
    return N

B=[4,2,3,5]
n=10

def prob(n,B,t): #on va regarder la probabilité de positionnement des bateaux
    M=np.zeros((n,n))
    for i in range (t):
        M+=grille(n,B)
    return M

M=prob(n,B,996)
plt.imshow(M,cmap='gray')
```

On voit sur le graphe que les bateaux se situent le plus souvent au centre de la grille.
On en déduit qu'il est plus intéressant de cibler le centre au début de la partie.

On va maintenant faire s'affronter deux grilles de bateaux et on veut recupérer le bateau qui gagne pour avoir une grille qui a le plus de chance de gagner.

```python
#monte carlo 

def test0(P): #on teste une matrice pour savoir si elle est nulle ou non
    n=len(P)
    k=0
    for i in range (n):
        for j in range (n):
            if P[i][j]==0:
                k+=1
    return (k==(n**2))


def fight(P,Q): #on se fait s'affronter deux grilles de bateaux et on retourne la grille gagnante
    n=len(P)
    F=np.zeros((n,n))
    R=np.zeros((n,n))
    S=np.zeros((n,n))
    k=0
    while not(test0(P)) and not(test0(Q)):
        verification = False
        while not(verification):
            x=randint(0,n-1)
            y=randint(0,n-1)
            if k%2==0:
                if R[x][y]==0:
                    verification = True
            else: 
                if S[x][y]==0:
                    verification = True
        if k%2==0:
            Q[x][y]=0
        else:
            P[x][y]=0
    vainqueur=0
    if test0(P):
        vainqueur=Q
    else:
        vainqueur=P
    return vainqueur

P=grille(n,B)
Q=grille(n,B)
print(fight(P,Q))

```

## Stratégie en damiers

On va montrer qu'attaquer la grille adversaire en diagonale permet d'être plus efficace.

On va prendre une grille de taille 10x10, et un bateau de 3 cases.
Donc pour un bateau à 3 cases on va attaquer 1 diagonale sur 3, comme sur cette image.



C'est-à-dire que comme le reste dans la division euclidienne de 10 par 3 est 1 on va avoir :

-pour les lignes où on commence à attaquer la première case on attaque au total 4 cases 

-pour les lignes où on commence à attaquer la deuxième case on attaque au total 3 cases

-pour les lignes où on commence à attaquer la troisième case on attaque au total 3 cases



Ainsi on aura 3 fois chacunes de ces lignes plus 1 que l'on supposera de la la forme 3 cases touchées pour garder un nombre minimal de cases attaquées.
Donc le nombre total de cases attaquées est : 33 

On attaque moins de la moitié des cases qui vaut 50 (100/2).

Ainsi grâce à la stratégie en damiers on remarque avec ce programme qu'en moins de 20 tentatives on peut toucher un bateau.

```python
'faire la liste de toutes les cases qu il faut toucher pour la stratégie en damier pour un bateau de taille k'
from math import*
def liste_attaques (k):
    r=10%k
    L=[]
    q=10//k
    for i in range (q):
        L.append([(i+1)*k,1]) #cibles de la première colonne
        x=(i+1)*k
        y=1
        while x>1 and 0<y<11:
            L.append([x-1,y+1]) #cibles en diagonale de cette colonne
            x=x-1
            y=y+1
    p=floor((10-k+r-1)/k)
    for i in range (p+1):
        L.append([10,k-r+1+i*k]) #cibles de la dernière ligne
        x=10
        y=k-r+1+i*k
        while x>1 and 0<y<10:
            L.append([x-1,y+1]) #cibles en diagonale de cette ligne
            x=x-1
            y=y+1
    return (L)

L2=liste_attaques(2)
L3=liste_attaques(3)
L4=liste_attaques(4)
L5=liste_attaques(5)

print(L3)
```
