## Analyse des Vecteurs Propres de la Matrice de Transition

La structure de la matrice $P$ est fondamentale pour comprendre ses vecteurs propres. Il s'agit d'une matrice stochastique par blocs, ce qui signifie que l'espace des états $\{1, \dots, 6\}$ est divisé en sous-ensembles aux comportements distincts : 
- l'état absorbant (1),
- les états transitoires (2, 3),
- et la classe terminale (4, 5, 6).

### 1. Les Valeurs Propres ($\lambda$)
Pour toute matrice de transition, la plus grande valeur propre est toujours $\lambda = 1$. Dans votre cas :
* **$\lambda = 1$ est une valeur propre double** : Cela indique la présence de deux classes fermées (ensembles d'états dont on ne peut plus sortir). Ici,
    * l'état $\{1\}$
    * et le bloc $C = \{4, 5, 6\}$.
* **$|\lambda| < 1$** : Les autres valeurs propres sont associées aux états transitoires. Elles déterminent la vitesse à laquelle la chaîne quitte les états 2 et 3 pour être absorbée par l'une des deux classes fermées.

### 2. Vecteurs Propres à Gauche (Mesures Invariantes)
Les vecteurs propres à gauche ($vP = \lambda v$) associés à $\lambda = 1$ représentent les distributions de probabilité stationnaires. Comme il y a deux classes fermées, il existe deux vecteurs de base linéairement indépendants :

#### A. Le point fixe de l'état 1
Si la chaîne commence ou tombe dans l'état 1, elle y reste indéfiniment. 
$$v_1 = (1, 0, 0, 0, 0, 0)$$

#### B. La loi stationnaire du bloc $C = \{4, 5, 6\}$
Le bloc $C$ est une classe terminale irréductible et apériodique. Une fois à l'intérieur, la chaîne circule entre 4, 5 et 6. À long terme, la distribution converge vers un équilibre interne. 


En résolvant le système $\mu_C P_C = \mu_C$ 
ou

$$
P_C = \begin{pmatrix} 
1/6 & 1/3 & 1/2 \\ 
1/2 & 0 & 1/2 \\ 
1/4 & 0 & 3/4 
\end{pmatrix}
$$

avec $\sum \mu_i = 1$, on obtient :
* $\mu_4 = 3/13 \approx 0.231$
* $\mu_5 = 1/13 \approx 0.077$
* $\mu_6 = 9/13 \approx 0.692$

```python


def simuler_trajectoire(matrice, x_init, n=100):
    """
    Simule une trajectoire de chaîne de Markov sans utiliser append.
    
    matrice : np.array (la matrice de transition)
    x_init  : int (état initial de 1 à N)
    n       : int (nombre de sauts/transitions)
    """
    # 1. On récupère la dimension dynamiquement
    nb_etats = matrice.shape[0]
    indices_etats = np.arange(1, nb_etats + 1)
    
    # 2. Pré-allocation du tableau (taille n + 1 pour inclure l'état initial)
    trajectoire = np.zeros(n + 1, dtype=np.int32)
    trajectoire[0] = x_init
    
    # 3. Boucle de simulation
    for t in range(1, n + 1):
        # On utilise l'état précédent (t-1) pour calculer l'état actuel (t)
        trajectoire[t] = np.random.choice(
            indices_etats, 
            p=matrice[trajectoire[t-1] - 1]
        )
        
    return trajectoire

# Exemple d'exécution
resultat = simuler_trajectoire(P, 6, n=100)
print(f"Trajectoire de {len(resultat)} points : {resultat}")
```

```python
# Matrice complète
P = np.array([
    [1,    0,    0,    0,    0,    0],
    [1/4, 1/2, 1/4, 0,    0,    0],
    [0,   1/5, 2/5, 1/5, 0,   1/5],
    [0,   0,   0,   1/6, 1/3, 1/2],
    [0,   0,   0,   1/2, 0,   1/2],
    [0,   0,   0,   1/4, 0,   3/4]
])

# Loi invariante théorique de C (calculée au préalable)
mu_C = np.array([3/13, 1/13, 9/13]) # Exact: [0.2308, 0.0769, 0.6923]

def simuler_Xn_naif(x_init, n,P, taille_echantillon=1000):
    nb_etats = P.shape[0]
    etats = np.arange(1, nb_etats + 1)
    final_states = []

    for _ in range(taille_echantillon):
        courant = x_init
        for _ in range(n):
            courant = np.random.choice(etats, p=P[courant - 1])
        final_states.append(courant)
    
    return np.array(final_states)

import numpy as np

def simuler_Xn_parallel(x_init, n, P, taille_echantillon=1000):
    """
    Simule l'état final Xn pour un grand échantillon de manière vectorisée.
    
    x_init : État de départ (1 à N)
    n : Nombre de pas de temps
    P : Matrice de transition (référence)
    taille_echantillon : Nombre de trajectoires simulées en parallèle
    """
    nb_etats = P.shape[0]
    # On travaille en indices 0 à N-1 pour NumPy
    etats_actuels = np.full(taille_echantillon, x_init - 1)
    
    # Pré-calcul des probabilités cumulées pour chaque ligne de P
    # Cela permet de transformer le tirage aléatoire en une recherche d'indice
    P_cum = np.cumsum(P, axis=1)
    
    for _ in range(n):
        # 1. Générer un nombre uniforme [0, 1) pour chaque trajectoire
        r = np.random.rand(taille_echantillon, 1)
        
        # 2. Trouver le prochain état
        # Pour chaque ligne i de P_cum correspondant à l'état actuel :
        # On cherche le premier indice j où r < P_cum[i, j]
        #this is basically the trick they use
        #https://github.com/numba/numba/issues/2539
        etats_actuels = (r > P_cum[etats_actuels]).sum(axis=1)
        
    # Retourner les états en repassant à la nomenclature {1...N}
    return etats_actuels + 1

# Paramètres du test
n_values = [5, 20, 50]
x_depart = 4 # On part d'un état dans C
etats_C = [4, 5, 6]

plt.figure(figsize=(15, 5))

for i, n in enumerate(n_values):
    echantillon = simuler_Xn_parallel(x_depart, n, P)
    
    # Calcul des probabilités empiriques pour 4, 5 et 6
    prob_empiriques = [np.mean(echantillon == s) for s in etats_C]
    
    plt.subplot(1, 3, i+1)
    x_axis = np.arange(len(etats_C))
    plt.bar(x_axis - 0.2, prob_empiriques, 0.4, label='Empirique', color='skyblue')
    plt.bar(x_axis + 0.2, mu_C, 0.4, label='Théorique (Invariante)', color='orange')
    
    plt.xticks(x_axis, etats_C)
    plt.title(f"n = {n}")
    plt.ylim(0, 1)
    if i == 0: plt.ylabel("Probabilité")
    plt.legend()

plt.suptitle(f"Convergence vers la loi invariante (Départ X0={x_depart})")
plt.tight_layout()
plt.show()
```

3. Estimer, sur la base d’un échantillon de taille 1000 de trajectoires de la chaîne, la probabilité d’absorption par 1 et par (C)
- en partant de $x = 2$
- puis de $x = 3$.
- Comparer ces valeurs aux valeurs théoriques vues en TD.



```python
import numpy as np

# Matrice de transition (P)
P = np.array([
    [1,    0,    0,    0,    0,    0],
    [1/4, 1/2, 1/4, 0,    0,    0],
    [0,   1/5, 2/5, 1/5, 0,   1/5],
    [0,   0,   0,   1/6, 1/3, 1/2],
    [0,   0,   0,   1/2, 0,   1/2],
    [0,   0,   0,   1/4, 0,   3/4]
])

def estimer_absorption(x_depart, 
                       n_sim=1000, 
                       n_pas=100):
    nb_etats = P.shape[0]
    etats = np.arange(1, nb_etats + 1)
    
    resultats_finaux = []
    
    for _ in range(n_sim):
        courant = x_depart
        for _ in range(n_pas):
            # Tirage de l'état suivant
            courant = np.random.choice(etats, p=P[courant - 1])
            
            # Optimisation : si on atteint une classe terminale, on s'arrête
            if courant == 1 or courant >= 4:
                break
        resultats_finaux.append(courant)
    
    resultats_finaux = np.array(resultats_finaux)
    
    # Calcul des proportions
    prob_1 = np.mean(resultats_finaux == 1)
    prob_C = np.mean(resultats_finaux >= 4)
    
    return prob_1, prob_C

# Exécution pour les deux points de départ
for x in [2, 3]:
    p1, pC = estimer_absorption(x)
    print(f"En partant de x = {x} :")
    print(f"  Probabilité d'absorption par l'état 1 : {p1:.3f}")
    print(f"  Probabilité d'absorption par le bloc C : {pC:.3f}\n")
```

```python
#complicated uses masks

def estimer_absorption_parallel(x_depart, P, 
                                n_sim=1000, n_pas=100):
    """
    Estime les probabilités d'absorption de manière vectorisée.
    """
    # Initialisation de tous les états à x_depart (indexé à 0 pour NumPy)
    etats_actuels = np.full(n_sim, x_depart - 1)
    
    # Pré-calcul des probabilités cumulées
    P_cum = np.cumsum(P, axis=1)
    
    # Masque pour suivre les trajectoires non encore absorbées
    # (État 1 est index 0, Bloc C est index >= 3)
    non_absorbe = np.ones(n_sim, dtype=bool)

    for _ in range(n_pas):
        # On ne met à jour que les trajectoires qui ne sont pas encore dans une classe terminale
        if not np.any(non_absorbe):
            break
            
        # Tirage aléatoire pour toutes les trajectoires
        r = np.random.rand(n_sim, 1)
        
        # Calcul du nouvel état pour tout le monde
        nouveaux_etats = (r > P_cum[etats_actuels]).sum(axis=1)
        
        # Mise à jour uniquement pour ceux qui n'étaient pas encore absorbés
        etats_actuels[non_absorbe] = nouveaux_etats[non_absorbe]
        
        # Mise à jour du masque : on s'arrête si l'état est 0 (état 1) ou >= 3 (états 4,5,6)
        non_absorbe = (etats_actuels != 0) & (etats_actuels < 3)

    # Conversion des index (0-5) vers les noms d'états (1-6)
    resultats_finaux = etats_actuels + 1
    
    # Calcul des proportions
    prob_1 = np.mean(resultats_finaux == 1)
    prob_C = np.mean(resultats_finaux >= 4)
    
    return prob_1, prob_C
```

### 3. Comparaison avec les valeurs théoriques

Pour valider les simulations, on utilise le calcul théorique des probabilités d'absorption. Soit $h_i$ la probabilité d'être absorbé par l'état **1** en partant de l'état $i$. Le système d'équations se base sur la décomposition au premier pas :

1.  **$h_1 = 1$** : L'état 1 est absorbant.
2.  **$h_2 = \frac{1}{4}h_1 + \frac{1}{2}h_2 + \frac{1}{4}h_3$** : Depuis 2, on va en 1 (prob. 1/4), on reste en 2 (prob. 1/2) ou on va en 3 (prob. 1/4).
3.  **$h_3 = \frac{1}{5}h_2 + \frac{2}{5}h_3 + \frac{2}{5}(0)$** : Depuis 3, on va en 2 (prob. 1/5), on reste en 3 (prob. 2/5) ou on entre dans le bloc $C$ (prob. 1/5 + 1/5 = 2/5). Si on entre dans $C$, la probabilité d'atteindre 1 devient nulle.

En résolvant ce système :
* De (3), on tire : $\frac{3}{5}h_3 = \frac{1}{5}h_2 \implies h_3 = \frac{1}{3}h_2$.
* En remplaçant dans (2) : $\frac{1}{2}h_2 = \frac{1}{4} + \frac{1}{4}(\frac{1}{3}h_2) \implies \frac{1}{2}h_2 - \frac{1}{12}h_2 = \frac{1}{4} \implies \frac{5}{12}h_2 = \frac{3}{12} \implies h_2 = \frac{3}{5}$.



#### Tableau récapitulatif des probabilités d'absorption

| État Initial ($x$) | Cible | Valeur Théorique | 
| :--- | :--- | :--- |
| **$x = 2$** | État 1 | **0,600** ($3/5$) | 
| **$x = 2$** | Bloc $C$ | **0,400** ($2/5$) 
| **$x = 3$** | État 1 | **0,200** ($1/5$) | 
| **$x = 3$** | Bloc $C$ | **0,800** ($4/5$) 

*Note : Les valeurs théoriques exactes dépendent de la résolution précise du système linéaire. Les résultats de la simulation (échantillon de 1000) doivent fluctuer autour de ces valeurs avec un écart type d'environ $\sqrt{p(1-p)/1000} \approx 0,015$.*
