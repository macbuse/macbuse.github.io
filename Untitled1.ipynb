```python
import matplotlib.pyplot as plt

# Data for the probability distribution of rolling two dice
sums = list(range(2, 13))
# probabilities = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36]


pp =  list(range(1,6))
pp.extend(list(range(6, 0, -1)))
probabilities = pp

# Convert probabilities to percentages
probabilities_percent = [p * 100 /36 for p in probabilities]

# Create the bar chart
plt.figure(figsize=(10, 5))
plt.bar(sums, probabilities_percent, 
        color='skyblue', edgecolor='black')

plt.title('Rolling 2 dice')
plt.savefig('dice.png')
```

```python
import numpy as np

# Simulate rolling 1000 six-sided dice many times
num_dice = 1000
num_trials = 10000  # Number of simulations
rolls = np.random.randint(1, 7, (num_trials, num_dice))  # Roll dice
sums_1000_dice = np.sum(rolls, axis=1)  # Sum of dice in each trial

# Plot histogram of sums
plt.figure(figsize=(10, 5))
plt.hist(sums_1000_dice, bins=50, color='mediumseagreen', edgecolor='black', density=True)

# Labels and title
plt.xlabel("Sum of 1000 Dice")
plt.ylabel("Relative Frequency")
plt.title("Distribution of Sums When Rolling 1000 Six-Sided Dice")

# Show the plot
#plt.show()
plt.savefig('dice_simulation.png')
```
