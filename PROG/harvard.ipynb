```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(n):
    if n == 0:
         return np.array([[0,1], [1,0]])
    #F_n-1 is not a valid variable name
    F_n_1 = F(n-1)
    F_n = np.zeros((2*F_n_1.shape[0]-1,2))
    F_n[::2,] = F_n_1
    F_n[1::2,] = F_n_1[:-1] + F_n_1[1:]
    return F_n

X = np.arange(-10,10)
Y = X.copy()
a, b = np.meshgrid(X,X)

fig, ax = plt.subplots()
fig.set_size_inches(10,10)
ax.set_aspect('equal')
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

ax.set_xlim(-10,9)
ax.set_ylim(-10,9)

ax.scatter(a,b);

xx = F(7)

for s in [np.array([1,1]), np.array([1,-1])]:
    xx = s*xx
    T = np.hstack([-xx, xx])
    X, Y = T[:,::2],  T[:,1::2]
    for a, b in zip(X,Y):
        plt.plot(a,b,'r')



fig.savefig('evenmore_geos.png')
```
