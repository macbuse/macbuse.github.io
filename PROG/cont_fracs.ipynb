```python
def ctd(x):
    ''' recursive continued fraction'''
    a,b = x
    if a % b ==  0: 
        return [a//b]
    L = ctd((b, a % b))
    L.append( a // b)
    return L
    
```

```python
N = 144 + 89
LL = [ (k, ctd((N,k))) for k in range(N//5 + 1, N - 3) ]
LL = [ (k, ctd((N,k))) for k in range(3, N - 3) ]
tt = [ (x,max(y)) for x,y in LL ]  
```
