With $f(t)=\sum_{n\in\mathbb Z} c_n e^{i2\pi n t}$ and
$c_n=\int_0^1 f(t)e^{-i2\pi n t}\,dt$, write 
- $\omega=2\pi n$
- $\alpha:=b-a.$


Then

$$
\boxed{\,c_0=\int_0^1 f(t)\,dt=a(c_2-c_1)+\frac{\alpha}{2}\,(c_2^2-c_1^2)\,}
$$

$$
\boxed{\;
c_n=\!\!\int_{c_1}^{c_2}(a+\alpha t)e^{-i\omega t}\,dt
=(e^{-i\omega c_2}-e^{-i\omega c_1})\!\left(\frac{\alpha}{\omega^2}-\frac{a}{i\omega}\right)
-\frac{\alpha}{i\omega}\,\big(c_2e^{-i\omega c_2}-c_1e^{-i\omega c_1}\big).}
$$







```python
a, b = -, 1 + 1J
c_1,c_2 = 0, 1
```

```python
def f_coeff(a,b,t_1,t_2,n):
    alpha = b - a
    if n == 0:
        return a*(t_2-t_1)+ alpha/2*(t_2**2-t_1**2)      
    omega = 2*np.pi*n
    w1, w2 = np.exp(-1J*np.array([t_1,t_2]))
    return (w2 - w1)*(alpha/omega**2 - a/1J/omega) - alpha/1J/omega*(t_2*w2 - t_1*w1)
```

```python
V = [f_coeff(0, 1 ,.25, .5,n) for n in range(-2,3)]
 
```

```python
T = np.linspace(0,1,40)
n = 2
M = 1J*np.zeros((5,40))
for n in range(-2,3):
    M[n+2] = np.exp(1J*2*np.pi*n*T)
Z = M.T @ V

plt.plot(Z.real, Z.imag)
```
