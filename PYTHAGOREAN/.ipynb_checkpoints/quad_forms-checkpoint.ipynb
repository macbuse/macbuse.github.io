```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def geo(a,b, sty = 'r'):
    R = abs(b-a)/2
    center = (a+b)/2
    T = np.linspace(0,np.pi,20)
    X = R*np.cos(T) + center
    Y = R*np.sin(T)
    plt.plot(X,Y,sty)
    return X,Y


def umbrella(x ,y,
             sty='b'):

    L = [x,y,-y,-x,x]
    for a, b in zip(L[1:],L):
        geo(a,b)

    geo(x,-y,sty=sty)
    geo(y,-x,sty=sty);

```

```python
import primefac

p = 7
ss = []
for n in  range(1,p):
    kk = p**2 - n**2
    facs = list(primefac.primefac(kk))
    print(kk, facs)
    uu = [(np.prod(facs),1)]
    for k in range(1,len(facs)):
        A,B = facs[:k], facs[k:]
        a,b = np.prod(A), np.prod(B)
        if a % 2 == 1 or b %2 == 1:
            continue
        uu.append((a,b))
    ss.append(uu)
ss
```

```python
def  mod_exp(a,b,m):
    if b == 0:
        return 1
    elif b%2 == 0:
        return mod_exp(a*a%m,b//2,m)
    else:
        return a*mod_exp(a,b-1,m)%m

```
