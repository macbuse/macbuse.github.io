```python

def xyz_generator(n):
    """
    Generates strings of length 2 up to n using 'x', 'y', 'z'.
    Constraint: No two adjacent characters can be the same.
    Starts with the initial seeds ['xy', 'xz'].
    """
    # Initial seeds
    current_layer = ["xy", "xz"]
    
    # Yield the starting strings first
    for s in current_layer:
        yield s
    
    # Generate subsequent lengths up to n
    for _ in range(n - 2):
        next_layer = []
        for s in current_layer:
            for char in "xyz":
                if s[-1] != char:
                    next_str = s + char
                    yield next_str
                    next_layer.append(next_str)
        current_layer = next_layer

target_length = 5
for result in xyz_generator(target_length):
        print(result)

```
