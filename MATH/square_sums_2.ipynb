```python
def gcde(a, b): 
    # Base Case 
    if a == 0 :  
        return b,0,1
             
    gcd,x1,y1 = gcde(b%a, a) 
     
    x = y1 - (b//a) * x1 
    y = x1 
     
    return gcd,x,y
```

```python
def gcd(a,b):
    x = []
    if b >a:
        a,b = b,a
    while b > 0:
        a, b, q = b, a % b, a // b
        x.append(q)
    return a, x
        
```
