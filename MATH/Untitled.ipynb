```python
import numpy as np

def Farey(n):
    F_n = np.array([[0,1], [1,1]],dtype=int)
    for k in range(n):
        tmp = np.zeros((2*F_n.shape[0]-1,2),dtype=int)
        # copy F_n in 
        tmp[::2,] = F_n
        # add medians in
        tmp[1::2,] = F_n[:-1] + F_n[1:]
        F_n = tmp
    return F_n

QQ = Farey(4)
```
