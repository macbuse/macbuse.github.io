Let 

$$Q := -9(x+y)^2 + 45(y-x)^2 + 10(x-z)^2$$

and we denote by $B$ the associated inner product:

$$B(u,v) := \frac14(Q(u+v) - Q(u-v)).$$

The quadratic form is preserved by a triple of involutions


$$
gx = \begin{pmatrix}
-1 & 19 & 19 \\ 
0 & 0 & 1 \\ 
0 & 1 & 0
\end{pmatrix},\,
gy = \begin{pmatrix}
0 & 0 & 1 \\ 
3/2 & -1 & 3/2 \\ 
1 & 0 & 0
\end{pmatrix},\,
gz = \begin{pmatrix}
0 & 1 & 0 \\ 
1 & 0 & 0 \\ 
1 & 1 & -1
\end{pmatrix}
$$

The product of the involutions is:

$$
gy.gz.gx = 
\begin{pmatrix}
-1 & 18 & 20 \\ 
-1/2 & 8 & 25/2 \\ 
0 & 0 & 1
\end{pmatrix}
$$

In fact it is easy to see that the product fixes a vector:


$$
\begin{pmatrix}
-1 & 18 & 20 \\ 
-1/2 & 8 & 25/2 \\ 
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
17\\3\\-1
\end{pmatrix}
= 
\begin{pmatrix}
17\\3\\-1
\end{pmatrix}
$$

We have a cycle:

$$\vec{v_0} : = \begin{pmatrix}
17\\3\\-1
\end{pmatrix}
\xrightarrow{gx}
\begin{pmatrix}
21\\-1\\3
\end{pmatrix}
\xrightarrow{gz}
\begin{pmatrix}
-1\\21\\17
\end{pmatrix}
\xrightarrow{gy}
\begin{pmatrix}
17\\3\\-1
\end{pmatrix}
= \vec{v_0}
$$

Now we can compute the (normalized) inner products:

$$\frac{B(v_i,v_j)}{\sqrt{B(v_i,v_i)B(v_j,v_j)}}$$

where $\vec{v}_1 = gx.\vec{v}_0, \vec{v}_2 = gz.gx.\vec{v}_0 = g_y.\vec{v}_0$

And the table get's usback to where we started from:



| geodesic 1 | geodesic 2  | inner product|
|----------------|--------------|-------|
| $v_0$          | $v_1$        | -3    |
| $v_1$          | $v_2$        | -21   |
| $v_2$          | $v_0$        | -17   |

# First approach

Calculate the hyperbolic distance between the fixed points 

```
fp_x = [19,1,1]
fp_y = [1,1.5,1]
fp_z = [1,1,1]
```

of the involutions ```gx,gy,gz``` that generate Doan's group:


one can do this using the inner product $B$ associated to the quadatic form :

$\cosh( d(u,v)) = B(u,v)$

$d(u,v) = 1/2 \ell_\gamma$ where $\gamma$ 
is the loxodromic obtained as a composition of the 
inversions in $a$ and $b$.


The length of the *companion* orthogeodesic $\gamma^*$ satisfies

$\cosh(1/2\ell_\gamma) = \sinh(1/4 \ell_\delta) \sinh(1/2 \ell_{\gamma^*})$

---

## Better way

$gx,gy,gz$ are elliptics and correspond to half turns around the midpoints of a triple of disjoint ortho geodesics.

the quotient of the hyperbolic plane by the group they generate is an orbifold which is a disc with a triple of cone points of cone angle $\pi$

the product ${gz}.{gy}.{gx}$ is loxodromic and corresponds to the holonomy of the boundary geodesic of the orbifold.

lifting to $\mathbb{H}$ we obtain a "hexagon" as in the figure below.

---

Now we can compute the ortho lengths of the dotted arcs in the hyperboloid formula
using the (normalized) inner product:

$$\cosh{\ell_{\gamma^*}} = \frac{B(u,v)}{\sqrt{B(u,u)B(v,v)}}$$

where $u,v$ are the vectors on the hyperboloid that represent a pair of
axes joined by the ortho geodesic (common perpendicular) $\gamma^*$.

![](../orbifold.svg)

[tt](./doan_fig.png)

````python
import numpy as np

def m2t(X):
    def l2t(R):
        return  ' & '.join([str(_) for _ in R])
    
    body = ' \\\\ \n'.join([l2t(_) for _ in X ])
    return f"\\begin{{pmatrix}}\n{body}\n\\end{{pmatrix}}"
````

````python
def Q(v):
    x,y,z =v
    return x**2 + 36*y**2 + 55*z**2 - 18*x*y - 20*x*z - 90*y*z

def ip(u,v):
    # standard way to get the inner product from a q-form
    uu = np.array(u)
    vv = np.array(v)

    return .25*(Q(uu+vv)-Q(uu-vv))

def DD(v):
    # auxilliary function in Doan
    x,y,z =v
    return x**2 + y**2 + z**2 + 2*x*y*z -1
````

## Important

so this says you get the original group back

````python
def Q(v):
    x,y,z =v
    return x**2 + 36*y**2 + 55*z**2 - 18*x*y - 20*x*z - 90*y*z

def ip(u,v):
    # standard way to get the inner product from a q-form
    uu = np.array(u)
    vv = np.array(v)
    return .25*(Q(uu+vv)-Q(uu-vv))

u = np.array([21,-1,3])
v = np.array([17,3,-1])

gy@gz@gx@v

L = v, gx@v, gz@gx@v, v
for x,y in zip(L,L[1:]):
    cosh = ip(x,y)/np.sqrt(Q(x) * Q(y) )
    print(x,y)
    print(cosh)
````

````python
loxo = gy@gz@gx
# find the eigenvectors on the light cone
fs,fu  = [_ for _ in np.linalg.eig(loxo)[1].T if abs(Q(_)) < .00001]
# randomly pick points in the plane
a, b = fs + fu, fs - fu
nn = np.cross(fs,fu)
# find which one is on the axis of loxo in the plane
pt = [_ for _ in [a,b] if Q(_) < 0][0]
# calculate the translation length
ip(loxo@pt,pt)/np.sqrt(Q(pt) * Q(loxo@pt) )
````
