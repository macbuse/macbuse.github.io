```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax  = plt.subplots(1, figsize=(10, 10))
ax.set_aspect('equal')

N = 7
for k in range(N):
    ax.plot([-k,k],[-1,1],'r')
    ax.plot([-k,k],[1,-1],'y')
    ax.plot([k/2,-k/2],[.5,-.5],'ro')
    ax.plot([-k/2,k/2],[.5,-.5],'yo')

ax.plot([-N,N],[.5]*2, 'b')
ax.plot([-N,N],[-.5]*2, 'b')

ax.grid(color='0.8')

ax.set_title('euclidean arcs');
```

```python
fig, ax  = plt.subplots(1, figsize=(10, 10))
ax.set_aspect('equal')

T = np.linspace(-np.pi/2,np.pi/2,20)
X = np.sin(T)

N = 7
for k in range(N):
    ax.plot(k*(X+1)/2,T/np.pi + .5,'r')
    ax.plot(-k*(X+1)/2,-T/np.pi -.5,'r')
    ax.plot(-k*(X+1)/2,T/np.pi + .5,'y')
    ax.plot(k*(X+1)/2,-T/np.pi -.5,'y')

ax.plot([-N,N],[.5]*2, 'b')
ax.plot([-N,N],[-.5]*2, 'b')

ax.grid(color='0.8')
ax.set_title('hyperbolic arcs');
```
