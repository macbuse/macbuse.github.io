```python
def geo(a,b, sty = 'r'):
    R = abs(b-a)/2
    center = (a+b)/2
    T = np.linspace(0,np.pi,20)
    X = R*np.cos(T) + center
    Y = R*np.sin(T)
    plt.plot(X,Y,sty)
    return X,Y


def umbrella(x ,y,
             sty='b'):

    L = [x,y,-y,-x,x]
    for a, b in zip(L[1:],L):
        geo(a,b)

    geo(x,-y,sty=sty)
    geo(y,-x,sty=sty);

```
