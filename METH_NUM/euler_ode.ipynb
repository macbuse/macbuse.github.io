# Euler's method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# Midpoint method

[on wikipedia](https://en.wikipedia.org/wiki/Midpoint_method)

- local error at each step  is of order
$ O\left(h^{3}\right)$, 
- giving a global error of order
$ O\left(h^{2}\right)$.

Thus, while more computationally intensive than Euler's method, 
the midpoint method's error generally decreases faster as $ h\to 0$.

The methods are examples of a class of higher-order methods known as Runge–Kutta methods.

```python

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Midpoint  Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) )   

plt.figure(figsize = (12, 8))

plt.plot(t, -np.exp(-t), 'r', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.plot(t, s, 'bo--', label='Approximate')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
# Define parameters t,y -> t
f = lambda t, s: s # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
    
euler = np.copy(s)

s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) ) 
    
midpoint = s

plt.figure(figsize = (12, 8))
plt.plot(t, np.exp(t), 'go--', label='Exact');
plt.plot(t, euler, 'bo--', label='Euler');
plt.plot(t, midpoint, 'ro--', label='midpoint');

plt.legend(loc='lower right');
```

- $(t,y) \mapsto ty$
- exact solution $\exp(t^2/2)$

$ y' = ty \Rightarrow y'/ y = t   \Rightarrow  \log y = t^2 $


```python
# Define parameters t,y -> ty

f = lambda t, s: t*s  # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
    
euler = np.copy(s)
s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) ) 
    
midpoint = s

plt.figure(figsize = (12, 8))
plt.plot(t, np.exp(t**2/2), 'go--', label='Exact');
plt.plot(t, euler, 'bo--', label='Euler');
plt.plot(t, midpoint, 'ro--', label='midpoint');

plt.legend(loc='lower right');
```

## more complicated


$ y' = y + \cos(t) \Rightarrow y'- y = \cos(t)   
\Rightarrow y = C\exp(t) + .5( \sin(t) - \cos(t) ) $

and $ C = y(0) + .5$.


```python
# Define parameters t,y -> ty

f = lambda t, s: s + np.cos(t)  # ODE
h = 0.1 # Step size
t = np.arange(0, 6 + h, h) # Numerical grid

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
    
euler = np.copy(s)
s = np.zeros(len(t))
s[0] = 1

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) ) 
    
midpoint = s

plt.figure(figsize = (12, 8))
plt.plot(t, (1 + .5)*np.exp(t) +  .5*( np.sin(t) - np.cos(t)), 'go--', label='Exact');
plt.plot(t, euler, 'bo--', label='Euler');
plt.plot(t, midpoint, 'ro--', label='midpoint');

plt.legend(loc='lower right');
```

## errors



```python


f = lambda t, s: t*s  # ODE

def euler(f, 
          y0=0, 
          subdivision=np.linspace(0,1,10)):
    
    # Explicit Euler Method
    t = subdivision
    s = np.zeros_like(subdivision)
    h = t[1] - t[0]
    s[0] = y0
    for i, y in enumerate(s[:-1]):
        s[i + 1] = y + h*f( t[i], y)
    return s

def midpoint(f,
             y0=0, 
             subdivision=np.linspace(0,1,10)):
    
    # Explicit Euler Metho
    t = subdivision
    s = np.zeros_like(subdivision)
    h = t[1] - t[0]
    s[0] = y0
    for i, y in enumerate(s[:-1]):
        s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) ) 
    return s

T = subdivision = np.linspace(0,1,10)

plt.figure(figsize = (12, 8))
plt.plot(T, np.exp(T**2/2), 'go--', label='Exact');
plt.plot(T, euler(f, y0=1, subdivision=T), 'bo--', label='Euler');
plt.plot(T, midpoint(f,y0=1, subdivision=T), 'ro--', label='midpoint');

plt.legend(loc='lower right');

```
