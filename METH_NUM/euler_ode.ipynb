# Euler's method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# Midpoint method

[on wikipedia](https://en.wikipedia.org/wiki/Midpoint_method)

- local error at each step  is of order
$ O\left(h^{3}\right)$, 
- giving a global error of order
$ O\left(h^{2}\right)$.

Thus, while more computationally intensive than Euler's method, 
the midpoint method's error generally decreases faster as $ h\to 0$.

The methods are examples of a class of higher-order methods known as Rungeâ€“Kutta methods.

```python

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Midpoint  Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f( t[i] + h/2 , s[i] + h/2*f(t[i],s[i]) )   

plt.figure(figsize = (12, 8))

plt.plot(t, -np.exp(-t), 'r', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.plot(t, s, 'bo--', label='Approximate')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
