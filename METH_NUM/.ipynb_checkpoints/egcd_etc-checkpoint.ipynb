```python
def dd(x):
    if x < 0 : 
        x += 11
    if x % 2 == 1: 
        x += 11
    return x

I = [ dd(egcd(_,11)[1]) for _ in range(1,11)]
```

```python
F = [ dd(-2*x % 11) for x in I]
```

```python
def L2R(L):
    return '& ' + ' & '.join([str(_) for _ in L]) + ' \\\\'
```

```python
print('\n'.join([ L2R(range(1,11)), L2R(I), L2R(F)] )) 
```

```python
import numpy as np

def Farey(n):
    F_n = np.array([[0,1], [1,1]])
    for k in range(n):
        tmp = np.zeros((2*F_n.shape[0]-1,2))
        # copy F_n in 
        tmp[::2,] = F_n
        # add medians in
        tmp[1::2,] = F_n[:-1] + F_n[1:]
        F_n = tmp
    return F_n

QQ = Farey(4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches(16,8)
ax.axis('equal')

T = np.linspace(0,2*np.pi, 40)
motif = np.exp(1J*T)/2

for vv in QQ:
    col = 'k'
    r = np.sqrt(2)/vv[1]**2
    if vv[0] % 2 == 1 : 
        r /= 2
        col = 'r'
    mm  = r*motif 
    plt.plot(mm.real + vv[0]/vv[1], 
             mm.imag + .5*r,col);
    plt.plot(mm.real + 2 -  vv[0]/vv[1], 
             mm.imag + .5*r,col);
```
